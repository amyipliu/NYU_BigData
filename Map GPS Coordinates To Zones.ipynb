```python
import geopandas as gpd
import pandas as pd
import rtree
from shapely.geometry import Point
data=pd.read_csv('/Volumes/Samsung USB/taxidata/2012/yellow_tripdata_2012-07.csv')
```

```python
#Load Map Shape File
zones = gpd.read_file('/Volumes/Samsung USB/taxidata/shapefiles/taxi_zones.shp').to_crs({'init' :'epsg:4326'})
zones['zone'] = zones.zone.astype('category')
zones['borough'] = zones.borough.astype('category')
zones.head()
```

```python
#Map Zones to Data
subset=data[['pickup_longitude','pickup_latitude']][:20] #Test a small sample
geometry = [Point(xy) for xy in zip(subset.pickup_longitude, subset.pickup_latitude)]
gf = subset.set_geometry(geometry, crs={'init' :'epsg:4326'})

gf2 = gpd.sjoin(gf, zones, how="inner")



```
